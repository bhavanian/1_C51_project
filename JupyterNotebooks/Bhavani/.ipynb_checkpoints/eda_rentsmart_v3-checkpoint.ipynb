```python
parcel_stats = pd.DataFrame(df.loc[df.violation_type == 'Housing Complaints'].groupby('parcel').violation_type.count())
```

```python
parcel_stats['units_res'] = df.groupby('parcel').units_res.sum()
```

```python
parcel_stats['housing_complaints_per_unit'] = parcel_stats['violation_type'] / parcel_stats['units_res']
```

```python
parcel_stats.replace([np.inf, -np.inf], np.nan, inplace=True)
parcel_stats = parcel_stats.dropna()
```

```python
parcel_stats.housing_complaints_per_unit.hist(bins=50)
plt.title('Histogram: Distribution of Housing Complaints Per Unit by Parcels')
plt.xlabel('Housing Complaints Per Unit')
plt.ylabel('Freq (Number of Parcels)')
plt.show()
```

```python
print('10 Parcels with the Most Housing Complaints per Residential Unit')
parcel_stats.housing_complaints_per_unit.nlargest(10)
```

```python
print('10 Parcels with the Least Housing Complaints per Residential Unit')
parcel_stats.housing_complaints_per_unit.nsmallest(10)
```

```python
complaint_cluster_feats_df = df[['violation_type','description']]
```

```python
complaint_cluster_feats_df = pd.get_dummies(complaint_cluster_feats_df)
```

```python
complaint_cluster_feats = StandardScaler().fit_transform(complaint_cluster_feats_df)
```

```python
t0 = time.perf_counter()
tsne_viz = TSNE(n_components=2, learning_rate='auto',
                  init='random').fit(complaint_cluster_feats)
print(time.perf_counter() - t0)
```

```python
plt.scatter(tsne_viz.embedding_[:,0],tsne_viz.embedding_[:,1])
plt.title('TSNE Visualization of Complaint Types')
plt.show()
```

```python
kmeans = KMeans(n_clusters=50, random_state=99).fit(complaint_cluster_feats)
```

```python
plt.scatter(tsne_viz.embedding_[:,0],tsne_viz.embedding_[:,1],c=kmeans.labels_)
plt.show()
```

```python
pca = PCA().fit(complaint_cluster_feats)
```

```python
complaint_PCA = pca.transform(complaint_cluster_feats)
```

```python
complaint_PCA.shape
```

```python
plt.scatter(complaint_PCA[:,0],complaint_PCA[:,1])
plt.show()
```
