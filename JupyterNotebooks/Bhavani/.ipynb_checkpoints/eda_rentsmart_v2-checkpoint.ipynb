```python
data_path = 'RentSmart/20220314_download/tmpdzll6bw0.csv'
```

```python
df = pd.read_csv(data_path, parse_dates=['date'], dtype={'zip_code':str})
```

```python
plt.scatter(df.latitude,df.longitude)
plt.show()
```

```python
property_cluster_feats = df.drop_duplicates(subset='address')[['neighborhood','zip_code',#'parcel',
                                                               'year built','year remodeled','property_type']]
```

```python
property_cluster_feats = property_cluster_feats.dropna()
```

```python
# property_cluster_feats['parcel'] = property_cluster_feats['parcel'].astype('category')
```

```python
property_cluster_feats = pd.get_dummies(property_cluster_feats)
```

```python
property_cluster_feats = StandardScaler().fit_transform(property_cluster_feats)
```

```python
property_cluster_feats.shape
```

```python
t0 = time.perf_counter()
tsne_viz = TSNE(n_components=2, learning_rate='auto',
                  init='random').fit(property_cluster_feats)
print(time.perf_counter() - t0)
```

```python
tsne_viz.embedding_.shape
```

```python
plt.scatter(tsne_viz.embedding_[:,0],tsne_viz.embedding_[:,1])
plt.show()
```

```python
property_cluster_feats_df = df.drop_duplicates(subset='address')[['latitude','longitude']]
property_cluster_feats_df = property_cluster_feats_df.dropna()
property_cluster_feats_df = property_cluster_feats_df.merge(df[['neighborhood','parcel']],left_index=True,how='left',right_index=True)
```

```python
# property_cluster_feats['parcel'] = property_cluster_feats['parcel'].astype('category')
```

```python
property_cluster_feats = StandardScaler().fit_transform(property_cluster_feats_df)
```

```python
property_cluster_feats.shape
```

```python
plt.scatter(property_cluster_feats[:,0],property_cluster_feats[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=100, random_state=99).fit(property_cluster_feats)
```

```python
kmeans.labels_
```

```python
plt.scatter(property_cluster_feats[:,0],property_cluster_feats[:,1], c=kmeans.labels_, cmap='tab10')
plt.show()
```

```python
plt.scatter(property_cluster_feats[:,0],property_cluster_feats[:,1], c=property_cluster_feats_df.neighborhood.astype('category').cat.codes,cmap='tab10')
plt.show()
```

```python
plt.scatter(property_cluster_feats[:,0],property_cluster_feats[:,1], c=property_cluster_feats_df.parcel.astype('category').cat.codes,cmap='prism')
plt.show()
```

```python
t0 = time.perf_counter()
tsne_viz = TSNE(n_components=2, learning_rate='auto',
                  init='random').fit(property_cluster_feats)
print(time.perf_counter() - t0)
```
